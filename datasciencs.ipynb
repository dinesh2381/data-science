```python

import numpy as np
import matplotlib.pyplot as plt

# 2) Vertical bar chart for products A, B, C, D with sales values
products = ["A", "B", "C", "D"]
sales = [10, 20, 30, 40]
plt.figure(figsize=(2,4))
plt.bar(products, sales)
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# 3) 2x2 grid of subplots containing 4 types of plots:
fig, axes = plt.subplots(2, 2, figsize=(10,8))

# Line
axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title("Line: sin(x)")

# Scatter
rng = np.random.default_rng(0)
xs = rng.uniform(0, 10, 50)
ys = rng.uniform(0, 10, 50)
axes[0,1].scatter(xs, ys)
axes[0,1].set_title("Scatter: random points")

# Bar
axes[1,0].bar(products, sales)
axes[1,0].set_title("Bar: product sales")

# Histogram
data_hist = rng.normal(loc=5, scale=2, size=200)
axes[1,1].hist(data_hist, bins=15)
axes[1,1].set_title("Histogram: sample normal data")

fig.suptitle("2x2 Subplot Grid: 4 plot types")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
# 4/6) Generate 200 random values from a normal distribution and plot a histogram
samples = rng.normal(loc=0, scale=1, size=200)
plt.figure(figsize=(6,4))
plt.hist(samples, bins=20, edgecolor='black')  # black edges as requested
plt.title("Histogram of 200 normal random values (20 bins)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
